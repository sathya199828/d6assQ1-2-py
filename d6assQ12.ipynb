create a bank account

```
class bankaccount():
  def __init__(self,ownername,Balance,deposit,withdraw):
    self.ownername = ownername
    self.Balance = Balance
    self.deposit = deposit
    self.withdraw = withdraw
    

  def getDetails(self):
      print("name",self.ownername)
      print("Balance",self.Balance)
      print("deposit",self.deposit)
      print("withdraw",self.withdraw)
      

pug = bankaccount("sat", 1500 ,1000,5000)

pug.getDetails()
```

volume and surfacearea of a cone

```
import math
pi= math.pi
def volume(r,h):
    return(1/3)*pi*r*r*h
def surfacearea(r,s):
    return pi*r*s*pi*r*r

radius = float(5)
height = float(12)
slatheight = float(13)

print(volume(radius,height))
print(surfacearea(radius,slatheight))
```
